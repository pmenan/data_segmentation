# Deuxième partie Segmentation RFM(Récence, Fréquence, Montant)


```python
good_data.head()
```

```python
#We only use 'Panier moyen', 'Recence' , 'Frequence', 'Achats de l\'année' now 
feature = ['Age Client','Panier moyen', 'Recence' , 'Frequence', 'Achats de l\'année']
data_nma = good_data[feature]
data_nma
```

```python
data_nma_v2 = data_nma.copy() 
data_nma_v2 = data_nma.copy() 
data_nma_v2['NMA * Panier moyen'] = np.where(data_nma['Frequence']> 0, (12/data_nma['Frequence'])*data_nma['Panier moyen'], 0*0)
```

```python
data_nma_v2
```

```python
#make copy
data_actif = data_nma_v2.copy()
data_actif['ACTIF'] = np.where(data_actif['Recence']>=12, "actif", "non_actif")
```

```python
data_actif
```

```python
#groupe by actif
data_actif.sort_values(by=['ACTIF'], inplace=True)
data_actif
```

```python
#make copy
#the result will be short by "ACTIF" and "À RELANCER" because "data_actif" is, already short by "ACTIF"
data_relance = data_actif.copy()
data_relance['À RELANCER'] = np.where(data_relance['Recence'] > data_relance['Frequence'], "À relancer", "RAS")
data_relance
```
